```python
import seaborn as sns
sns.set_style("whitegrid")
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
mpl.rcParams['figure.dpi'] = 200
corr = df.select_dtypes(include = ['float64', 'int64']).iloc[:, :].corr()
plt.figure(figsize=(10, 10))
ax = sns.heatmap(corr, vmax=1, square=True)
# ax.set_xticklabels(rotation=30)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
```

```python
mpl.rcParams['figure.dpi'] = 250
plt.figure(figsize=(60, 60))
f, axes = plt.subplots(1, 2)
sns.boxplot(y=df["manufacturing cost"], ax=axes[0])
sns.boxplot(y=df["retail price"], ax=axes[1])
plt.subplots_adjust(wspace=1)
```

```python
sns.countplot(x = 'manufacturing cost', data = df)
```

```python
sns.distplot(df[df['retail price'].notnull()]['retail price'], kde=False, bins=10);
# sns.distplot(df['retail price'], kde=False, bins=10);
```

```python
df[df['retail price'].notnull()].describe()
```

```python
mpl.rcParams['figure.dpi'] = 250
plt.figure(figsize=(100, 100))
df_nona= df[df['retail price'].notnull()]
g = sns.pairplot(df_nona[['manufacturing cost', 'retail price']] )
```
